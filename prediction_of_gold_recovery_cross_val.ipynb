### Проверка правильности расчёта эффективности обогащения

```python
# Формирование датафрейма со значениями содержания
# золота в черновом концентрате (эффект обогащения)
rougher_output_recovery = pd.DataFrame(data=train['rougher.output.recovery'])
```

```python
# Проверочный расчёт и вывод на экран значений содержания
# золота в черновом концентрате (эффект обогащения)
rougher_output_recovery['cheked_rougher.output.recovery'] = (
    (train['rougher.output.concentrate_au'] * (train['rougher.input.feed_au'] - train['rougher.output.tail_au'])) / 
    (train['rougher.input.feed_au'] * (train['rougher.output.concentrate_au'] - train['rougher.output.tail_au']))
) * 100
rougher_output_recovery
```

```python
# Расчёт и вывод на экран средней абсолютной ошибки между значениями
# в 'rougher_output_recovery' и 'cheked_rougher.output.recovery'
mean_absolute_error(
    rougher_output_recovery['rougher.output.recovery'],
    rougher_output_recovery['cheked_rougher.output.recovery']
)
```

```python
# Вывод на экран датафрейма с отличающимися
# значениями эффекта обогащения
rougher_output_recovery[
    rougher_output_recovery['rougher.output.recovery'] !=
    rougher_output_recovery['cheked_rougher.output.recovery']
]
```

```python
# Список колонок для заполнения медианным значением
columns_list = [
    'primary_cleaner.input.xanthate',
    'primary_cleaner.state.floatbank8_a_air',
    'primary_cleaner.state.floatbank8_b_air',
    'primary_cleaner.state.floatbank8_c_air',
    'primary_cleaner.state.floatbank8_d_air',
    'rougher.input.floatbank11_xanthate',
    'rougher.state.floatbank10_e_air',
    'secondary_cleaner.state.floatbank2_a_air',
    'secondary_cleaner.state.floatbank2_b_air',
    'secondary_cleaner.state.floatbank3_a_air',
    'secondary_cleaner.state.floatbank4_a_air'
]
```

```python
# Создание функции для заполнения пропусков медианным
# значением с группировкой по году, месяцу и дню
# 1. Объявление функции drop_nan()
def drop_nan(
    data,
    col_date,
    col_fill,
    func_app
):
    # 2. Приведение типа данных в 'datetime64'
    data[col_date] = data[col_date].astype('datetime64')
    # 3. Создание дополнительных колонок с
    # годом, месяцем и днём недели
    data['year'] = data['date'].dt.year
    data['month'] = data['date'].dt.strftime('%A')
    data['day'] = data['date'].dt.strftime('%B')
    # 4. Создание сводной таблицы с медианным
    # значением по году, месяцу и дню
    data_pivot = data.pivot_table(
        index=['year', 'month', 'day'],
        values=col_fill,
        aggfunc=func_app
    )
    # 5. Объединение датафреймов data и data_pivot
    data_join = data.join(
        data_pivot,
        on=['year', 'month', 'day'],
        rsuffix='_median'
    )
    # 6. Создание переменной со списком для заполнения
    # названиями колонок, содержащих медианные значения
    col_list_median = []
    # 7. Цикл перебора названий колонок с добавлением
    # к ним суффикса '_median' и заполнение списка
    # col_list_median
    for i in col_fill:
        col_list_median.append(i + '_median')
    # 8. Цикл заполнения пропусков медианным значением
    for j in range(len(col_fill)):
        data_join.loc[
            data_join[col_fill[j]].isna(),
            col_fill[j]
        ] = data_join[col_list_median[j]]
    # 9. Декомпозиция датафрейма data_join
    data_join.drop(
        columns=['year', 'month', 'day'] + col_list_median,
        axis=1,
        inplace=True
    )
    # 10. Возвращение функции
    return data_join
```

## Обучение и проверка модели

### Создание функции, вычисляющей значение sMAPE

```python
# Создание функции, вычисляющей значение симметричного
# среднего абсолютного процентного отклонения (sMAPE)
# 1. Объявление функции smape_find()
def smape_find(
    y_true,
    y_pred
):
    # 2. Расчёт значения sMAPE
    smape_result = 100 * sum(
        np.abs(y_true - y_pred) /
        ((np.abs(y_true) + np.abs(y_pred)) / 2)
    ) / len(y_true)
    # 3. Возвращение функции
    return smape_result
```

### Создание метрики качества sMAPE

```python
# Создание метрики качества sMAPE для аргумента
# 'scoring' в функции cross_val_score
smape_scorer = make_scorer(
    smape_find,
    greater_is_better=False
)
```

### Создание функции, вычисляющей итоговое sMAPE

```python
# Создание функции, вычисляющей итоговое значение симметричного
# среднего абсолютного процентного отклонения (sMAPE)
# 1. Объявление функции final_smape()
def final_smape(
    first_model_name,
    second_model_name,
    X_features,
    y_true_first,
    y_true_second,
    rate_first,
    rate_second
):
    # 2.1. Формирование предсказаний первой моделью
    y_pred_first = first_model_name.predict(X_features)
    # 2.2. Формирование предсказаний второй моделью
    y_pred_second = second_model_name.predict(X_features)
    # 3.1. Расчёт первого значения sMAPE
    first_smape = 100 * sum(
        np.abs(y_true_first - y_pred_first) /
        ((np.abs(y_true_first) + np.abs(y_pred_first)) / 2)
    ) / len(y_true_first)
    # 3.2. Расчёт второго значения sMAPE
    second_smape = 100 * sum(
        np.abs(y_true_second - y_pred_second) /
        ((np.abs(y_true_second) + np.abs(y_pred_second)) / 2)
    ) / len(y_true_second)
    # 4. Расчёт финального значения sMAPE
    final_smape_result = rate_first * first_smape + rate_second * second_smape
    # 5. Возвращение функции
    return final_smape_result
```

### Создание функции, формирующей обученные модели

```python
# Создание функции regression_select(), обучающей модели для
# задач регрессии с перебором параметров, кросс-валидацией
# и выводом метрик качества
# 1. Объявление функции regression_select() и задание
# параметров данной функции
def regression_select(
    model_name,
    features_train_income,
    target_train_income,
    depth_treshold=None,
    estimators_treshold=None,
    estimators_step=None,
    scoring_income=None,
    cv_income=5,
    best_cvs=50
):
    # A.1. Логическое условие для отбора и обучения
    # модели решающего дерева
    if model_name == DecisionTreeRegressor:
        # A.2.1. Создание переменной, содержащей список списков для
        # заполнения значениями максимальной глубины дерева,
        # минимального количества выборок в листьях, количества блоков
        # кросс-валидации, метрик качества
        data_value = []
        # A.3.2. Создание переменной со списком названий колонок
        columns_name = [
            'depth_value',
            'leaf_value',
            'number_cv_blocks',
            'smape_cross_mean'
        ]
        # A.3. Цикл перебора значений максимальной
        # глубины дерева
        for depth in range(1, depth_treshold):
            # A.4. Цикл перебора значений минимального
            # количества выборок в листьях
            for leaf in range(1, depth_treshold):
                # A.5. Цикл перебора количества блоков
                # кросс-валидации
                for block in range(2, cv_income):
                    # A.6.1. Создание объекта структуры данных
                    model = model_name(
                    max_depth=depth,
                    min_samples_leaf=leaf,
                    random_state=777
                    )
                    # A.6.2. Запуск кросс-валидации и формирование
                    # массива оценок выбранной метрики качества
                    cvs = cross_val_score(
                        model,
                        features_train_income,
                        target_train_income,
                        cv=block,
                        scoring=scoring_income
                    )
                    # A.6.3. Вычисление среднего значения
                    # полученной метрики качества
                    mean_cvs = np.abs(cvs.mean())
                    # A.6.4. Запуск обучения модели на обучающей выборке
                    model.fit(features_train_income, target_train_income)
                    # A.6.5. Заполнение списка в переменной data_value
                    # значениями максимальной глубины дерева, минимального
                    # количества выборок в листьях, количества блоков
                    # кросс-валидации, метрик качества
                    data_value.append(
                        [
                            depth,
                            leaf,
                            block,
                            mean_cvs
                        ]
                    )
                    # A.7.1. Логическое условие сохранения параметров модели
                    # с лучшим значением mean_cvs
                    if mean_cvs < best_cvs:
                        # A.7.2. Сохранение текущего значения mean_cvs
                        best_cvs = mean_cvs
                        # A.7.3. Сохранение лучшей обученной модели
                        best_model = model
        # A.8.1. Формирование датафрейма с комбинациями максимальной глубины
        # дерева, минимального количества выборок в листьях, количества блоков
        # кросс-валидации, полученными значениями метрик качества
        new_data = pd.DataFrame(data=data_value, columns=columns_name)
        # A.8.2. Формирование датафрейма с лучшим значением mean_cvs
        df_cvs = new_data.loc[
            new_data['smape_cross_mean'] == new_data['smape_cross_mean'].min()
        ].reset_index(drop=True)
        # A.9. Вывод на экран лучшего значения mean_cvs
        # и значений параметров обучения модели
        print()
        print(
            'Лучшее среднее значение симметричного среднего абсолютного процентного отклонения (sMAPE): ', df_cvs['smape_cross_mean'][0], ' при значениях параметров:', '\n',
            '- максимальная глубина дерева: ', df_cvs['depth_value'][0], '\n',
            '- минимальное количество выборок в листе: ', df_cvs['leaf_value'][0], '\n',
            '- количество блоков кросс-валидации: ', df_cvs['number_cv_blocks'][0],
            sep=''
        )
        print()
        # A.10. Вывод сформированного датафрейма new_data на экран
        print()
        print('Датафрейм со значениями параметров обучения и метриками качества:')
        display(new_data)
        print()
        # A.11. Вывод модели решающего дерева на экран
        print('Визуализация модели решающего дерева:')
        plt.figure(figsize=(75, 75))
        tree.plot_tree(best_model)
        plt.show()
        # A.12. Возвращение функции
        return best_model
# --------------------------------------------------------------------------------------------------------
    # B.1. Логическое условие для отбора и обучения
    # модели случайного леса
    if model_name == RandomForestRegressor:
        # B.2.1. Создание переменной, содержащей список списков для
        # заполнения значениями количества деревьев в лесу, 
        # максимальной глубины дерева, минимального количества
        # выборок в листьях, метрик качества
        data_value = []
        # B.2.2. Создание переменной со списком названий колонок
        columns_name = [
            'number_tree',
            'depth_value',
            'leaf_value',
            'number_cv_blocks',
            'smape_cross_mean'
        ]
        # B.3. Цикл перебора значений максимальной глубины дерева
        for estim in range(1, estimators_treshold, estimators_step):
            # B.4. Цикл перебора значений максимальной глубины дерева
            for depth in range(1, depth_treshold):
                # B.5. Цикл перебора значений минимального
                # количества выборок в листьях
                for leaf in range(1, depth_treshold):
                    # B.6. Цикл перебора количества блоков
                    # кросс-валидации
                    for block in range(2, cv_income):
                        # B.7.1. Создание объекта структуры данных
                        model = model_name(
                            n_estimators=estim,
                            max_depth=depth,
                            min_samples_leaf=leaf,
                            random_state=777
                        )
                        # B.7.2. Запуск кросс-валидации и формирование
                        # массива оценок выбранной метрики качества
                        cvs = cross_val_score(
                            model,
                            features_train_income,
                            target_train_income,
                            cv=block,
                            scoring=scoring_income
                        )
                        # B.7.3. Вычисление среднего значения
                        # полученной метрики качества
                        mean_cvs = np.abs(cvs.mean())
                        # B.7.4. Запуск обучения модели на обучающей выборке
                        model.fit(features_train_income, target_train_income)
                        # B.7.5. Заполнение списка в переменной data_value
                        # значениями количества деревьев в лесу, максимальной
                        # глубины дерева, минимального количества выборок в
                        # листьях, количества блоков кросс-валидации, метрик
                        # качества
                        data_value.append(
                            [
                                estim,
                                depth,
                                leaf,
                                block,
                                mean_cvs
                            ]
                        )
                        # B.8.1. Логическое условие сохранения параметров модели
                        # с лучшим значением mean_cvs
                        if mean_cvs < best_cvs:
                            # B.8.2. Сохранение текущего значения mean_cvs
                            best_cvs = mean_cvs
                            # B.8.3. Сохранение лучшей обученной модели
                            best_model = model
        # B.9.1. Формирование датафрейма с комбинациями количества деревьев
        # в лесу, максимальной глубины дерева, минимального количества выборок
        # в листьях, количества блоков кросс-валидации, полученными значениями
        # метрик качества
        new_data = pd.DataFrame(data=data_value, columns=columns_name)
        # B.9.2. Формирование датафрейма с лучшим значением mean_cvs
        df_cvs = new_data.loc[
            new_data['smape_cross_mean'] == new_data['smape_cross_mean'].min()
        ].reset_index(drop=True)
        # B.10. Вывод на экран лучшего значения mean_cvs
        # и значений параметров обучения модели
        print()
        print(
            'Лучшее среднее значение симметричного среднего абсолютного процентного отклонения (sMAPE): ', df_cvs['smape_cross_mean'][0], ' при значениях параметров:', '\n',
            '- количество деревьев в лесу: ', df_cvs['number_tree'][0], '\n',
            '- максимальная глубина дерева: ', df_cvs['depth_value'][0], '\n',
            '- минимальное количество выборок в листе: ', df_cvs['leaf_value'][0], '\n',
            '- количество блоков кросс-валидации: ', df_cvs['number_cv_blocks'][0],
            sep=''
        )
        print()
        # B.11. Вывод сформированного датафрейма new_data на экран
        print()
        print('Датафрейм со значениями параметров обучения и метриками качества:')
        display(new_data)
        # B.12. Возвращение функции
        return best_model
# --------------------------------------------------------------------------------------------------------
    # C.1. Логическое условие для отбора и обучения
    # модели линейной регрессии
    if model_name == LinearRegression:
        # C.2.1. Создание переменной, содержащей список списков для
        # заполнения значениями количества блоков кросс-валидации,
        # метрик качества
        data_value = []
        # C.2.2. Создание переменной со списком названий колонок
        columns_name = [
            'number_cv_blocks',
            'smape_cross_mean'
        ]
        # C.3. Цикл перебора количества блоков
        # кросс-валидации
        for block in range(2, cv_income):
            # C.4.1. Создание объекта структуры данных
            model = model_name()
            # C.4.2. Запуск кросс-валидации и формирование
            # массива оценок выбранной метрики качества
            cvs = cross_val_score(
                model,
                features_train_income,
                target_train_income,
                cv=block,
                scoring=scoring_income
            )
            # C.4.3. Вычисление среднего значения
            # полученной метрики качества
            mean_cvs = np.abs(cvs.mean())
            # C.4.4. Запуск обучения модели на обучающей выборке
            model.fit(features_train_income, target_train_income)
            # C.4.5. Заполнение списка в переменной data_value
            # значениями количества блоков кросс-валидации,
            # метрик качества
            data_value.append(
                [
                    block,
                    mean_cvs
                ]
            )
            # C.5.1. Логическое условие сохранения параметров модели
            # с лучшим значением mean_cvs
            if mean_cvs < best_cvs:
                # C.5.2. Сохранение текущего значения mean_cvs
                best_cvs = mean_cvs
                # C.5.3. Сохранение лучшей обученной модели
                best_model = model
        # C.6.1. Формирование датафрейма с комбинациями количества блоков
        # кросс-валидации, полученными значениями метрик качества
        new_data = pd.DataFrame(data=data_value, columns=columns_name)
        # C.6.2. Формирование датафрейма с лучшим значением mean_cvs
        df_cvs = new_data.loc[
            new_data['smape_cross_mean'] == new_data['smape_cross_mean'].min()
        ].reset_index(drop=True)
        # C.7. Вывод на экран лучшего значения mean_cvs
        # и значений параметров обучения модели
        print()
        print(
            'Лучшее среднее значение симметричного среднего абсолютного процентного отклонения (sMAPE): ', df_cvs['smape_cross_mean'][0], ' при значениях параметров:', '\n',
            '- количество блоков кросс-валидации: ', df_cvs['number_cv_blocks'][0],
            sep=''
        )
        print()
        # C.8. Вывод сформированного датафрейма new_data на экран
        print()
        print('Датафрейм со значениями параметров обучения и метриками качества:')
        display(new_data)
        # C.9. Возвращение функции
        return best_model
```
